```python
passing_grade = 60
eval_list=[]
for i in grades:
    if i>=passing_grade:
        eval_list.append('Pass')
    else:
        eval_list.append('Fail')
eval_list
```

```python
passing_grade = 60
eval_list=[]
for i in grades:
  eval_list.append(i>=passing_grade)
eval_list
```

```python
passing_grade = 60
eval_list = ['Pass' if i>=passing_grade else 'Fail' for i in grades]
eval_list
```

5. The grades of students in Hydroinformatics are as follows; in order to take this course you need to have “B-” or higher in “advanced water resources management”. Evaluate their scores to see how many are eligible to take this course.


```python
grades = ['B+', 'B', 'D', 'F', 'D-', 'C+', 'C-', 'B+', 'A+', 'F']
```

```python
eligable_students = 0
for i in grades:
	if i=='B-' or i=='B' or i=='B+' or i=='A-' or i=='A' or i=='A+':
		eligable_students += 1
eligable_students
```

```python
passing_grades = ['B-', 'B', 'B+', 'A-', 'A', 'A+']
status = [1 if i in passing_grades else 0 for i in grades]
status
```

```python
eligable_students = 0
for i in status:
  eligable_students += i
eligable_students
```

```python
eligable_students = sum(status)
eligable_students
```
